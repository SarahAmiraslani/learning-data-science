### Introduction to R
- `c()` - used to create a vector of numbers, c for *concatenate*. 
- `length()` - used to check the length of a set
- `ls()` - allows us to look at a list of all of the objects, such as data and functions, that we have saved so far. 
- `rm()` - used to remove any saved objects that we do not want. 
    - It is possible to remove all objects at once using `rm(list=ls())
    
- `seq()` - used to create a sequence of numbers. E.g., `seq(a,b)` makes a vector of integers between `a` and `b`. 
    - `seq(0,1,length=10)` makes a sequence of 10 numbers that are equally spaced betwen 0 and 1. 
    - Note that typing `3:11` is shorthand for `seq(3,11)` for integer arguments.

```r
# join together the numbers 1,3,2, and 5 and save them as a vector names x
x = c(1,3,2,5)
y = c(1,4,3,6)

x
y

x+y
```

```r
seq(1,10)
seq(0,1,length=5)
```

- `matrix()` - can be used to create a matrix of numbers. Inputs include the `data` (the entries of the matrix), the number of rows (`nrow`), and the number of columns (`ncol`).
- `dim()` - outputs the number of rows followed by the number of columns of a given matrix

```r
x = matrix(data=c(1,2,3,4),nrow=2,ncol=2)
x
```

- `sqrt()`- returns the square root of each element of a vector or matrix.
- `^2` - raises each element of the vector or matrix to the power of 2 (any powers are possible)

```r
sqrt(x)
x^2
```

- `rnorm()` - generates a vector of random normal variables with a mean of 0 and standard deviation of 1 by defualt. However, the mean and standard deviation can be altered using the `mean` and `sd` arguments. 
    - If we want our code to reproduce the exact same set of random numbers, we can use the `set.seed()` function which takes an aribitrary integer argument. 
    
- `cor()` - computes the correlation between two vectors of numbers.
- `mean()` - computes the mean of a vector of numbers. 
- `var()` - computes the variance of a vector of numbers. 
- `sd()` - computes the standard deviation of a vector of numbers. 

```r
x = rnorm(50)
y = x + rnorm(50,mean=50,sd=.1)

cor(x,y)
```

#### Indexing Data
Suppose our data is stored in the matrix `A`. 
- `A[row_num,col_num]` can be used to index
- We can also select multiple rows and columns at a time by providing vectors as the indices. `A[c(row_num,row_num,...),c(col_num,col_num,...)]` or `A[row_num_start:row_num_end,col_num_start:col_num_end]`
- If no index for the columns or no index for the rows is provided, this indicates that R should include all columns or all rows. 

```r
A = matrix(1:16,4,4)
A
```

```r
A[2,3]
```

```r
A[c(2,3),c(2,4)]
```

```r
A[1:3 ,2:4]
```

```r
A[1:2,]
```

The use of a negative sign `-` in the index tells R to keep all rows or columns
except those indicated in the index.

```r
A[-c(1,3) ,]
```

#### Graphics
- `plot()` is the primary way to plot data in R, 
    - `plot(x,y)` produces a scatterplot of the numbers in `x` versus the numbers in `y`
    - other arguments of `plot()`: 
        - `xlab`: x label
        - `ylab`: y label
        - `main`: title
- `hist()` - plot a histogram. 
- `pairs()` - creates a scatterplot matrix for every pair of variables for a given data set. 
- `contour()` produces a contour plot in order to represent three-dimensional data
- `image()` produces a color-coded plot (i.e., heatmap) whose colors depend on the z-value
- `persp()` can be used to produce a three-dimensional plot where the arguments `theta` and `phi` control the angles at which the plot is viewed. 

```r
hist(Auto$mpg,breaks=20)
```

```r
pairs(Auto)
```

```r
x=rnorm(100)
y=rnorm(100)
plot(x,y,xlab=" this is the x-axis",ylab=" this is the y-axis",main=" Plot of X vs Y")
```

```r
x=seq(-pi ,pi ,length =50)
y=x
f=outer(x,y,function (x,y)cos(y)/(1+x^2))
contour (x,y,f)
```

```r
fa=(f-t(f))/2
image(x,y,fa)
```

```r
persp(x,y,fa)
```
